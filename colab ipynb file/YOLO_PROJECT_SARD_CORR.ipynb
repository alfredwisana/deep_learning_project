```python
# Function to perform object detection on an image and save the result
def detect_and_save(model, image_path, output_dir):
    # Load image
    img = cv2.imread(image_path)

    # Perform detection
    results = model(img)
    result = results[0]  # Assuming single result for simplicity

    # Draw bounding boxes and labels on the image
    img_with_boxes = result.plot()

    # Save the image with bounding boxes
    filename = os.path.basename(image_path)
    output_path = os.path.join(output_dir, filename)
    cv2.imwrite(output_path, img_with_boxes)

    print(f"Processed: {filename}")

# Path to the directory containing images
input_dirs = [
    '/content/drive/MyDrive/project2/dataset/test/test/filtered_image/snow',
    '/content/drive/MyDrive/project2/dataset/test/test/filtered_image/rain',
    '/content/drive/MyDrive/project2/dataset/test/test/filtered_image/fog',
]

# Output directory to save processed images
output_dir = '/content/drive/MyDrive/project2/videos/output/corr_images_after_train'
os.makedirs(output_dir, exist_ok=True)

# Iterate over each directory
for input_dir in input_dirs:
    # Iterate over each image in the directory
    for filename in os.listdir(input_dir):
        if filename.endswith('.jpg') or filename.endswith('.png'):
            image_path = os.path.join(input_dir, filename)
            detect_and_save(model, image_path, output_dir)

print("All images processed and saved.")
```
